```python
def count_word_frequency(text):
    text = text.lower()
    for znak in [".", ",", "!", "?"]:
        text = text.replace(znak,"")
    words = text.split()
    
    frequency = {}
    for word in words:
        frequency[word] = frequency.get(word, 0) + 1
    
    return frequency
```

```python
text = "Это просто пример. Просто очень простой пример текста для примера."
frequency_dict = count_word_frequency(text)
print(frequency_dict)
```

```python
def calculate_price(price, discount=0, tax_rate=0.2):
    price = price * (1-discount/100)
    final = price * (1 + tax_rate)
    return final
```

```python
calculate_price(250) #только налог 0.2
```

```python
calculate_price(250, 5)
```

```python
calculate_price(250, tax_rate=0.5)
```

```python
calculate_price(250, tax_rate=0.5, discount = 3)
```

```python
calculate_price(price = 250, tax_rate=0.5, discount = 3)
```

```python
from datetime import datetime

def analyze_log(file_path):
    try:
        with open(file_path, "r", encoding="utf-8") as f:
            counts = {"INFO": 0, "WARNING": 0, "ERROR": 0}
            min_date = datetime.strptime("3000-01-30 00:00:00", "%Y-%m-%d %H:%M:%S")
            max_date = datetime.strptime("1900-01-30 00:00:00", "%Y-%m-%d %H:%M:%S")
            for line in f:
                try:
                    parts = line.strip().split(',', 2)
                
                    #print(parts)
                    date_str, level, message = parts
                    
                    if level in counts:
                        counts[level] += 1
                        
                    date_time = datetime.strptime(date_str, "%Y-%m-%d %H:%M:%S")
                    if date_time < min_date:
                        min_date = date_time
                    if date_time > max_date:
                        max_date = date_time

                except (IndexError, ValueError):
                    continue
            time_range = (min_date.strftime("%Y-%m-%d %H:%M:%S"),max_date.strftime("%Y-%m-%d %H:%M:%S"))
           
            return {'counts': counts, 'time_range': time_range}
        #print(f"Данные успешно прочитаны")
    except FileNotFoundError:
        return {}                       
```

```python
result = analyze_log('log.txt')
print(result)                 
```
