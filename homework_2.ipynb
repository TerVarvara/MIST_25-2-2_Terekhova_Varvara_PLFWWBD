Задача 1

```python
num = int(input('Введите число:'))
for i in range (1, num+1):
    if i%3 == 0 and i%5 == 0:
        print('FizzBuzz')
    elif i%3 ==0:
        print('Fizz')
    elif i%5 ==0:
        print('Buzz')
    else:
        print(i)
```

Задача 2

```python
shopping_list = []
while True:
    option = int(input('Выберите пункт меню:\n1. Добавить\n2. Показать\n3. Выйти\n'))
    if option == 1:
        prod = input('Введите название продукта ')
        shopping_list.append(prod)
    elif option == 2:
        if len(shopping_list)==0:
            print('Список покупок пуст')
        else:
            for index, name in enumerate(shopping_list):
                print(f"{index + 1}. {name}")
    elif option == 3:
        print('Программа завершена')
        break
    else: print('Неизвестная команда, попробуйте снова.')
```
